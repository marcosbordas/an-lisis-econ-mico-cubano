## Conclusión del proyecto

### El análisis realizado a partir de tres fuentes distintas y cinco gráficos evidencia una realidad contundente: el costo de la cesta básica y de los productos esenciales para adultos mayores supera ampliamente la pensión mínima de **3056 CUP**. La brecha entre ingresos y gastos no es solo numérica, sino que se traduce en una dificultad cotidiana para cubrir necesidades básicas de alimentación, higiene y salud.

## Posibles vías de solución

### - **Revisión y actualización periódica de la pensión mínima**: vincularla al costo real de la canasta básica, garantizando que cubra al menos las necesidades esenciales.
### - **Políticas de subsidio focalizado**: otorgar apoyo directo en productos críticos (alimentos, medicamentos, artículos de higiene) para adultos mayores y personas vulnerables.
### - **Mayor regulación y transparencia en los canales de comercialización**: reducir la dispersión de precios y asegurar una oferta estable en MIPYMES y mercados estatales.
